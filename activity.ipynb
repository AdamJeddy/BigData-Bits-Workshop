```python
def read_document(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.readlines()

def map_function(document):
    tokens = document.strip().split()
    return [(token.lower(), 1) for token in tokens] # converting tokens to lowercase
```

```python
documents = read_document('text_10k.txt')
print(documents)
```

```python
# Map Phase  - Apply the map function to each document
mapped_results = []

for document in documents:
    mapped_results.extend(map_function(document))
```

```python
documents = read_document('text_10k.txt')

# Map Phase  - Apply the map function to each document
mapped_results = []

for document in documents:
    mapped_results.extend(map_function(document))

# Oraganize mapped results by key word to prepare for reduce phase
intermediate = {}
for key, value in mapped_results:
    if key not in intermediate:
        intermediate[key] = []
    intermediate[key].append(value)

# Reduce Funtion: Aggregates count for each word
def reduce_function(items):
    word, counts = items[0], items[1]
    return (word, sum(counts))

# reduce phase
reduce_results = map(reduce_function, intermediate.items())

# print(list(reduce_results))
```

```python
word_counts = list(reduce_results)
most_common = max(word_counts, key=lambda x: x[1])
least_common = min(word_counts, key=lambda x: x[1])
avg = sum(count for _, count in word_counts) / len(word_counts)
print(avg)

print(f"Most common wordL {most_common}")
print(f"Least common wordL {least_common}")
```
